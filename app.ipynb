Usar as 'libs'

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Ler a base e organiza-la

```python
dados = pd.read_csv('fatura.csv')
dados
```

Iniciar o projeto de analise



* Valor total da fatura

```python
valor_total_cartao = dados['ValorCompra'].sum()
print(f"O valor total da fatura e: R$ {valor_total_cartao}")
```

Valor gastor por pessoa

```python
# Criar varios blocos organizacoes, pegando a planilha e fazendo bloco por bloco
valor_gasto_pessoa = dados.groupby('Nome')['ValorCompra'].sum().reset_index().sort_values(by='ValorCompra', ascending=False)
valor_gasto_pessoa
```

Valor gasto por categoria

```python
# Criar bloco por bloco de analise 
valor_gasto_categoria = dados.groupby('categoria')['ValorCompra'].sum().reset_index().sort_values(by='ValorCompra', ascending=False)
valor_gasto_categoria
```

Dias que mais ocorreram gastos

```python

gastos_dia = dados.groupby('data_compra')['ValorCompra'].sum().reset_index().sort_values(by='data_compra')
gastos_dia 
```

Gastor por pessoa (graficos)

```python
plt.figure(figsize=(10,6))
plt.bar(valor_gasto_pessoa['Nome'], valor_gasto_pessoa['ValorCompra'], color='red')
plt.title("Valor gasto por Pessoa")
plt.xlabel('Titular')
plt.ylabel('Valor gasto (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 
```

Grafico de gastos por Categoria

```python
plt.figure(figsize=(10,6))
plt.barh(valor_gasto_categoria['categoria'], valor_gasto_categoria['ValorCompra'])
plt.xlabel('Valor Gasto')
plt.ylabel('Categoria')
plt.show()
```

Graficos dos gastos por Pessoa


```python
plt.figure(figsize=(5,5))
plt.pie(valor_gasto_pessoa['ValorCompra'], labels=valor_gasto_pessoa['Nome'], autopct='%1.1f%%')
plt.title("Gastos por Pessoa")
plt.show()
```
